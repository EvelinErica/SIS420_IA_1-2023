```python
class Nodo:
    def __init__(self, estado, hijo=None):
        self.estado = estado
        self.hijo = None
        self.padre = None
        self.cost = None
        self.set_hijo(hijo)
        #def set_estado(self, estado):
        #self.estado = estado

    def get_estado(self):
        return self.estado


    def set_hijo(self, hijo):
        self.hijo = hijo
        if self.hijo is not None:
            for s in self.hijo:
                s.padre = self

    def get_hijo(self):
        return self.hijo

    def get_padre(self):
        return self.padre

    def set_padre(self, padre):
        self.padre = padre

    def set_estado(self, estado):
        self.estado = estado

    def get_destado(self):
        return self.estado

    def set_cost(self, cost):
        self.cost = cost

    def get_cost(self):
        return self.cost

    def equal(self, Nodo):
        if self.get_estado() == Nodo.get_estado():
            return True
        else:
            return False

    def on_list(self, lista_nodos):
        listed = False
        for n in lista_nodos:
            if self.equal(n):
                listed = True
        return listed

    def __str__(self):
        return str(self.get_estado())
```

```python
import random

def busqueda_BPPR(nodo_inicial, solution, visitado):
    visitado.append(nodo_inicial.get_estado())
    if nodo_inicial.get_estado() == solution:
        return nodo_inicial
    else:
        # Expandir nodos sucesores (hijos)
        hijo_nodo = nodo_inicial.get_estado().copy()
        for i in range (1,len(hijo_nodo)): #Para iterar sobre los Ã­ndices de una secuencia
          for j in range (0,len(hijo_nodo)-1):
            if hijo_nodo[j]>hijo_nodo[j+1]:
              temp = hijo_nodo[j]
              hijo_nodo[j]= hijo_nodo[j+1]
              hijo_nodo[j+1]=temp
              hijo = hijo_nodo
              print("nodo hijo ")
              print(hijo_nodo)
              new_hijo = Nodo(hijo)
              nodo_inicial.set_hijo([new_hijo])
           #   new_son.set_father(init_node)

        for hijo_nodo in nodo_inicial.get_hijo():
            if not hijo_nodo.get_estado() in visitado and heuristica(nodo_inicial, hijo_nodo):
                # Llamada recursiva
                solutn = busqueda_BPPR(hijo_nodo, solution, visitado)
                if solutn is not None:
                    return solutn
        return None


def heuristica(padre_node, hijo_nodo):
    padre_quality = 0
    hijo_quality = 0
    padre_data = padre_node.get_estado()
    hijo_data = hijo_nodo.get_estado()

    for n in range(1, len(padre_data)):
        if padre_data[n] > padre_data[n - 1]:
            padre_quality = padre_quality + 1
        if hijo_data[n] > hijo_data[n - 1]:
            hijo_quality = hijo_quality + 1

    if hijo_quality >= padre_quality:
        return True
    else:
        return False


if __name__ == "__main__":
    initial_state = []
    solution_state = []
    number_elements= int(input("introdusca el numero de elementos"))

    for i in range(0,number_elements):
      initial_state.append(i)
      solution_state.append(i)
    initial_state.reverse()
    random.shuffle(initial_state)
    print("nodo creado aleatoriamente")
    print(initial_state)
    print("nodo resultado")
    print(solution_state)
    visited_nodes = []
    nodo_inicial = Nodo(initial_state)
    solution_node = busqueda_BPPR(nodo_inicial, solution_state, visited_nodes)

    result = []
    node = solution_node
    while node.get_padre() is not None:
        result.append(node.get_estado())
        node = node.get_padre()
    result.append(initial_state)
    result.reverse()
    #print(result)
```
